```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier, XGBRegressor
from lightgbm import LGBMRegressor,LGBMClassifier
from catboost import CatBoostRegressor, CatBoostClassifier
from sklearn.svm import SVC,SVR
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.ensemble import VotingClassifier,VotingRegressor,StackingClassifier,StackingRegressor
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
X, y = load_breast_cancer(return_X_y=True)
print(X.shape)
print(y.shape)
print(X[:5])
pd.DataFrame(y).value_counts()
```

```python
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits=5)
```

```python
def classification_model(model):
    scores = cross_val_score(model, X, y, cv=kfold)
    return scores.mean()
```

```python
classification_model(XGBClassifier())
```

```python
classification_model(XGBClassifier(booster='gblinear'))
```

```python
classification_model(XGBClassifier(booster='dart'))
```

```python
classification_model(RandomForestClassifier(random_state=2))
```

```python
classification_model(LogisticRegression(max_iter=10000))
```

```python
classification_model(LinearDiscriminantAnalysis())
```

```python
classification_model(QuadraticDiscriminantAnalysis())
```

```python
classification_model(SVR())
```

```python
classification_model(SVR(kernel='rbf'))
```

```python
classification_model(LGBMClassifier())
```

```python
classification_model(CatBoostClassifier())
```

```python
# 위에서 돌린 기법들의 Ensemble을 얻기 위함
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
def y_pred(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_pred, y_test)
    print(score)
    return y_pred
```

```python
y_pred_gbtree = y_pred(XGBClassifier())
```

```python
y_pred_dart = y_pred(XGBClassifier(booster='dart'))
```

```python
y_pred_forest = y_pred(RandomForestClassifier(random_state=2))
```

```python
y_pred_logistic = y_pred(LogisticRegression(max_iter=10000))
```

```python
y_pred_lda=y_pred(LinearDiscriminantAnalysis())
```

```python
y_pred_qda=y_pred(QuadraticDiscriminantAnalysis())
```

```python
y_pred_lgb=y_pred(LGBMClassifier())
```

```python
y_pred_cat=y_pred(CatBoostClassifier())
```

```python
# 모든 기법에 따른 predicted output 비교
df_pred = pd.DataFrame(data= np.c_[y_pred_gbtree, y_pred_dart, y_pred_forest, y_pred_logistic, y_pred_lda,
                                  y_pred_qda,y_pred_lgb, y_pred_cat], 
                  columns=['gbtree', 'dart', 'forest', 'logistic', 'LDA', 'QDA','lgb','cat'])
df_pred.head()
```

```python
df_pred.corr()  # correlation 비교
# QDA와 LGB의 correlation은 0.97이므로 둘 중 하나 제거
# Logistic과 Catboost의 correlation은 0.97이므로 둘 중 하나 제거
```

```python
estimators = []
logistic_model = LogisticRegression(max_iter=10000)
estimators.append(('logistic', logistic_model))

xgb_model = XGBClassifier()
estimators.append(('xgb', xgb_model))

rf_model = RandomForestClassifier(random_state=2)
estimators.append(('rf', rf_model))

lda_model=LinearDiscriminantAnalysis()
estimators.append(('lda', lda_model))

# qda_model=QuadraticDiscriminantAnalysis()
# estimators.append(('qda', lda_model))

lgb_model=LGBMClassifier()
estimators.append(('lgb',lgb_model))

# cat_model=CatBoostClassifier()
# estimators.append(('cat',cat_model))
ensemble = VotingClassifier(estimators)  # 모든 estimator의 ensemble로 모델 돌리기
scores = cross_val_score(ensemble, X, y, cv=kfold)
print(scores.mean())
```

```python
base_models = []
base_models.append(('xgb', XGBClassifier()))
base_models.append(('rf', RandomForestClassifier(random_state=2)))
base_models.append(('lda', LinearDiscriminantAnalysis()))
base_models.append(('lgb', LGBMClassifier()))
base_models.append(('cat',CatBoostClassifier()))

# define meta learner model
meta_model = LogisticRegression()

# define the stacking ensemble
clf = StackingClassifier(estimators=base_models, final_estimator=meta_model)
scores = cross_val_score(clf, X, y, cv=kfold)
print(scores.mean())
```
